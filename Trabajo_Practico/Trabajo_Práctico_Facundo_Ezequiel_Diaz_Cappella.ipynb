#Modelo
- ¿Como represento el espacio de soluciones?
- ¿Cual es la función objetivo?
- ¿Como implemento las restricciones?


## Respuesta
- Representación del espacio de soluciones
    - **Cada calendario es una combinacion de:**
      - Partidos (20 equipos)
      - Franjas horarias para cada partido
      - Restricción de partidos en viernes y lunes
</br>
- Función objetivo
  - **Maximizar la audiencia total considerando:**
    - Categoría de los equipos
    - Franja horaria del partido
    - Reducción por coincidencias de horarios
$$ \text{maximize} \sum_{i=1}^{n} A(e1_i,e2_i) \cdot T(h_i) \cdot [1 - R(c_i)] $$
</br>
Variables:
</br>
$A(e1_i,e2_i)$ = audiencia base del partido entre equipos $e1_i$ y $e2_i$ </br>
$T(h_i)$ = coeficiente temporal de la franja horaria $h_i \in {V20,S12,...,L20}$</br>
$R(c_i)$ = reducción por $c_i$ coincidencias en franja $h_i$</br>
$n$ = número total de partidos</br>
$i$ = índice del partido</br>
</br>
- Restricciones
    - **Restricciones duras:**
      - Un partido el viernes - **Proponemos (equipos B y C)**
      - Un partido el lunes - **Proponemos (equipos B y C)**
      - Número total de partidos
      - Ningún equipo juega contra sí mismo
    - **Restricciones blandas:**
      - Variedad de franjas horarias
      - Reducción de audiencia por coincidencias
      - Distribución equilibrada de horarios
</br>
- Con este modelo buscamos optimizar la audiencia total mientras respetamos las restricciones del problema.

#Análisis
- ¿Que complejidad tiene el problema?. Orden de complejidad y Contabilizar el espacio de soluciones

## Respuesta
- Complejidad computacional:
    - $O(n!)$ donde n es el número de partidos posibles.
- Espacio de soluciones:
    - 20 equipos = 10 partidos por jornada
    - 10 franjas horarias disponibles
    - Dos partidos fijos (viernes y lunes)

- Cálculo:
    - Partidos posibles: $\binom{20}{2} = 190$ combinaciones de equipos
    - Para cada partido (excepto viernes/lunes): 8 franjas horarias posibles
    - Para viernes y lunes: **propenemos solo equipos B vs C**

 - Total predicible:
$$ \binom{11 \cdot 6}{2} \cdot 1 \cdot \binom{(11-1)(6-1)}{2} \cdot 1 \cdot 8^8 $$
Resulta en un espacio de soluciones del orden de $10^{12}$ posibilidades, por esto debemos buscar soluciones acordes (complejidad computacional alta), en este caso optamos por algoritmos genéticos.

#Diseño
- ¿Que técnica utilizo? ¿Por qué?

## Respuesta
La creación de este calendario de partidos de fútbol que maximiza la audiencia es un problema combinatorio complejo. Como vimos al calcular el espacio de soluciones. Hay muchas combinaciones posibles de partidos y franjas horarias, y encontrar la mejor solución de forma exhaustiva requiere un cálculo computacional muy alto que no es feasible.

Por esto optamos por usar los algoritmos genéticos, ya que nos otorgan:
- **Adaptabilidad**: Estos algoritmos son buenos para explorar espacios de busqueda grandes y complejos. Podemos agregar diferentes restricciones y objetivos como: **las categorías de los equipos**, **las franjas horarias** y **la reducción de audiencia por coincidencias**.

- **Flexibilidad**: No podemos garantizar encontrar la solución optima. Pero podemos acercarnos a ella con buenas soluciones y en un tiempo razonable además de sin grandes costos computacionales.
</br>
Comenzamos con una población aleatoria de soluciones, las evaluamos, filtramos las mejores. Luego las combinamos y mutamos para crear nuevas soluciones, y así sucesivamente. Este proceso iterativo permite encontrar soluciones que maximizan la audiencia del calendario de partidos.

```
## Solución
import random
import numpy as np

# Configuración de datos
EQUIPOS = {
    'A': ['Real Madrid', 'Barcelona', 'Atlético de Madrid'],
    'B': ['Sevilla', 'Real Sociedad', 'Valencia', 'Athletic Club', 'Betis',
          'Villarreal', 'Espanyol', 'Granada', 'Alavés', 'Levante', 'Osasuna'],
    'C': ['Celta', 'Mallorca', 'Eibar', 'Leganés', 'Valladolid', 'Getafe']
}

AUDIENCIA_BASE = {
    ('A', 'A'): 2.0, ('A', 'B'): 1.3, ('A', 'C'): 1.0,
    ('B', 'B'): 0.9, ('B', 'C'): 0.75, ('C', 'C'): 0.47
}

FRANJAS_HORARIAS = [
    'V20', 'S12', 'S16', 'S18', 'S20',
    'D12', 'D16', 'D18', 'D20', 'L20'
]

REDUCCION_COINCIDENCIA = {
    0: 0.0, 1: 0.25, 2: 0.45, 3: 0.60,
    4: 0.70, 5: 0.75, 6: 0.78, 7: 0.80, 8: 0.80
}

def obtener_categoria(equipo):
    for categoria, equipos in EQUIPOS.items():
        if equipo in equipos:
            return categoria
    raise ValueError(f"Equipo {equipo} no encontrado")

def calcular_audiencia_partido(equipo1, equipo2, franja):
    cat1, cat2 = obtener_categoria(equipo1), obtener_categoria(equipo2)
    base = AUDIENCIA_BASE.get(tuple(sorted((cat1, cat2))), 0)

    coeficientes = {
        'V20': 0.4, 'S12': 0.55, 'S16': 0.7, 'S18': 0.8, 'S20': 1.0,
        'D12': 0.45, 'D16': 0.75, 'D18': 0.85, 'D20': 1.0, 'L20': 0.4
    }
    return base * coeficientes[franja]

def evaluar_calendario(calendario):
    audiencia_total = 0
    conteo_franjas = {}

    # Contar coincidencias de franjas
    for _, _, franja in calendario:
        conteo_franjas[franja] = conteo_franjas.get(franja, 0) + 1

    for equipo1, equipo2, franja in calendario:
        audiencia_base = calcular_audiencia_partido(equipo1, equipo2, franja)

        # Reducir audiencia por coincidencias
        coincidencias = conteo_franjas[franja]
        factor_reduccion = 1 - REDUCCION_COINCIDENCIA.get(coincidencias, 0)

        audiencia_total += audiencia_base * factor_reduccion

    return audiencia_total

def crear_individuo(todos_equipos):
    """Crea un calendario aleatorio válido"""
    # Aseguramos partido en viernes y lunes con categorías diferentes

    equipos_disponibles = todos_equipos.copy()

    calendario = []

    # Viernes: equipo B vs equipo C. Proponemos estas categorías para los viernes
    equipos_b = [e for e in equipos_disponibles if obtener_categoria(e) == 'B']
    equipos_c = [e for e in equipos_disponibles if obtener_categoria(e) == 'C']

    if not equipos_b or not equipos_c:
        return None

    equipo_b = random.choice(equipos_b)
    equipos_b.remove(equipo_b)
    equipo_c = random.choice(equipos_c)
    equipos_c.remove(equipo_c)

    calendario.append((equipo_b, equipo_c, 'V20'))
    equipos_disponibles.remove(equipo_b)
    equipos_disponibles.remove(equipo_c)

    # Lunes: otro equipo B vs otro equipo C. Proponemos estas categorías para los lunes
    if not equipos_b or not equipos_c:
        return None

    equipo_b = random.choice(equipos_b)
    equipos_b.remove(equipo_b)
    equipo_c = random.choice(equipos_c)
    equipos_c.remove(equipo_c)

    calendario.append((equipo_b, equipo_c, 'L20'))
    equipos_disponibles.remove(equipo_b)
    equipos_disponibles.remove(equipo_c)

    # Rellenar resto de partidos
    while equipos_disponibles:
        if len(equipos_disponibles) < 2:
            return None

        equipo1 = random.choice(equipos_disponibles)
        equipos_disponibles.remove(equipo1)
        equipo2 = random.choice(equipos_disponibles)
        equipos_disponibles.remove(equipo2)

        franja = random.choice(FRANJAS_HORARIAS)
        calendario.append((equipo1, equipo2, franja))

    return calendario

def generar_poblacion_inicial(num_individuos, todos_equipos):
    poblacion = []
    intentos = 0
    while len(poblacion) < num_individuos and intentos < 1000:
        individuo = crear_individuo(todos_equipos)
        if individuo:
            poblacion.append(individuo)
        intentos += 1

    if len(poblacion) == 0:
        raise ValueError("No se pudo generar población inicial")

    return poblacion

def seleccion_padres(poblacion, fitness, num_padres):
    """Selecciona los mejores individuos como padres"""
    indices_ordenados = np.argsort(fitness)[::-1]
    return [poblacion[i] for i in indices_ordenados[:num_padres]]

def cruzar_calendarios(padre1, padre2):
    """Realiza cruce entre dos calendarios"""
    punto_cruce = random.randint(1, len(padre1)-1)
    hijo = padre1[:punto_cruce] + padre2[punto_cruce:]
    return hijo

def mutar_calendario(calendario, probabilidad_mutacion=0.1):
    """Muta un calendario"""
    if random.random() < probabilidad_mutacion:
        # Cambiar franja horaria de un partido aleatorio
        indice = random.randint(0, len(calendario)-1)
        calendario[indice] = (*calendario[indice][:2], random.choice(FRANJAS_HORARIAS))
    return calendario

def algoritmo_genetico(num_generaciones=100, tam_poblacion=50):
    """Algoritmo genético para optimizar calendario"""
    todos_equipos = [equipo for categoria in EQUIPOS.values() for equipo in categoria]

    # Generar población inicial
    poblacion = generar_poblacion_inicial(tam_poblacion, todos_equipos)

    mejor_audiencia = 0
    mejor_calendario = None

    for generacion in range(num_generaciones):
        # Evaluar fitness de la población
        fitness = [evaluar_calendario(calendario) for calendario in poblacion]

        # Encontrar el mejor calendario de esta generación
        mejor_actual = max(fitness)
        if mejor_actual > mejor_audiencia:
            mejor_audiencia = mejor_actual
            mejor_calendario = poblacion[fitness.index(mejor_actual)]

        # Selección de padres
        padres = seleccion_padres(poblacion, fitness, max(2, tam_poblacion // 2))

        # Generar nueva población
        nueva_poblacion = []
        while len(nueva_poblacion) < tam_poblacion:
            padre1 = random.choice(padres)
            padre2 = random.choice(padres)
            hijo = cruzar_calendarios(padre1, padre2)
            hijo = mutar_calendario(hijo)
            nueva_poblacion.append(hijo)

        poblacion = nueva_poblacion

    return mejor_calendario, mejor_audiencia

def ejecutar_optimizacion_repetida(num_intentos=100):
    mejor_audiencia = 0
    mejor_calendario = None
    mejor_intento = 0

    for intento in range(1, num_intentos + 1):
        calendario_actual, audiencia_actual = algoritmo_genetico()

        if audiencia_actual > mejor_audiencia:
            mejor_audiencia = audiencia_actual
            mejor_calendario = calendario_actual
            mejor_intento = intento

    print(f"\nMejor resultado en el intento {mejor_intento} de un total de 100 ejecuciones:")
    for partido in mejor_calendario:
        print(f"Partido: {partido[0]} vs {partido[1]}, Franja Horaria: {partido[2]}")
    print(f"\nAudiencia Total: {mejor_audiencia:.2f} millones")

    return mejor_calendario, mejor_audiencia, mejor_intento


ejecutar_optimizacion_repetida()
```
